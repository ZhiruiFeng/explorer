```python
# Add libraries
%load_ext autoreload
%autoreload 2
import os, sys
lib_path = os.path.abspath(os.path.join('../'))
sys.path.append(lib_path)

from env.grid import GCCPGrid
from search.mygraph import GridGraph
from search.explorer import Explorer

filename = "../data/grid/arena.map"
```

```python
Grid = GCCPGrid(filename)
G = GridGraph(Grid)
G.print_terrain()
```

```python
start = (3, 1)
goal = (32, 35)
explorer = Explorer(G)
if G.is_legal_goal(start, goal):
    # Dijkstra, get the optimal solution
    path, cost, number_of_visited, number_of_expanded = explorer.plan_dijkstra(start,goal)
    print("\nShortest path cost: {}".format(cost))
    print("Number of visited nodes: {}".format(number_of_visited))
    print("Number of expanded nodes: {}".format(number_of_expanded))
    explorer.print_exploration()
```

```python
if G.is_legal_goal(start, goal):
    # A* with unadmissible heuristic
    path, cost, number_of_visited, number_of_expanded = explorer.plan_astar(start,goal,'Manhattan')
    print("\nShortest path cost: {}".format(cost))
    print("Number of visited nodes: {}".format(number_of_visited))
    print("Number of expanded nodes: {}".format(number_of_expanded))
    explorer.print_exploration()
```

```python
if G.is_legal_goal(start, goal):
    # A* with admissible heuristic
    path, cost, number_of_visited, number_of_expanded = explorer.plan_astar(start,goal,'Straight')
    print("\nShortest path cost: {}".format(cost))
    print("Number of visited nodes: {}".format(number_of_visited))
    print("Number of expanded nodes: {}".format(number_of_expanded))
    explorer.print_exploration()
```
