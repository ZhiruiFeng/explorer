- Test on grig graph for game map

```python
# Add libraries
%load_ext autoreload
%autoreload 2
import os, sys
lib_path = os.path.abspath(os.path.join('../'))
sys.path.append(lib_path)

from env.grid import GCCPGrid
from search.mygraph import GridGraph
from search.dijkstra import dijkstra
from search.astar import astar

filename = "../data/grid/arena.map"
```

```python
Grid = GCCPGrid(filename)
G = GridGraph(Grid)
G.print_terrain()
```

```python
start = (3, 1)
goal = (32, 35)
if G.is_legal_goal(start, goal):
    # Dijkstra, get the optimal solution
    G.reset()
    path, cost = dijkstra(G, start, goal)
    print("\nShortest path cost: {}".format(cost))
    print("Number of visited nodes: {}".format(G.number_of_labeled()))
    print("Number of expanded nodes: {}".format(G.number_of_scanned()))
    G.mark_path(path)
    G.print_exploration(path)
    
    # A* with unadmissible heuristic
    G.reset()
    path, cost = astar(G, start, goal, 'Manhattan')
    print("\nShortest path cost: {}".format(cost))
    print("Number of visited nodes: {}".format(G.number_of_labeled()))
    print("Number of expanded nodes: {}".format(G.number_of_scanned()))
    G.mark_path(path)
    G.print_exploration(path)
    
    # A* with admissible heuristic 
    G.reset()
    path, cost = astar(G, start, goal, 'Straight')
    print("\nShortest path cost: {}".format(cost))
    print("Number of visited nodes: {}".format(G.number_of_labeled()))
    print("Number of expanded nodes: {}".format(G.number_of_scanned()))
    G.mark_path(path)
    G.print_exploration(path)
```
